```python
import numpy as np
from numpy import transpose, trace, multiply, power, dot
from numpy.linalg import multi_dot, matrix_power
import scipy.stats as ss
from scipy.special import comb
import pandas as pd
import time
from tqdm import tqdm
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
import data_gen as dg
import stat_test as st
```

# Asymptotic null distribution of degree test

```python
# Set seed
np.random.seed(seed=13)
```

```python
# Data generation parameters
n = 200
alpha = 0.1
```

```python
# Test parameters
N_rep = 500
level = 0.05
```

```python
# Empirical null distribution
test_stats = []
for i in tqdm(range(N_rep)):
    time.sleep(0.01)
    # Sample an adjacency matrix under the null
    A = dg.sample_null_adj_mat(n, alpha)
    # Compute degree test statistic
    t = st.degree_test(A, level)['test_stat']
    test_stats.append(t)
```

```python
# Plot histogram
degree_test = pd.DataFrame({'test_stat': test_stats})
ax = sns.histplot(data=degree_test, x='test_stat', stat='density')
xx = np.arange(-4,+4,0.001)
norm_pdf = ss.norm.pdf(xx)
ax.plot(xx,norm_pdf, 'r', lw=2)
# Save figure to file
plt.savefig('experiment1_nulldist_degree.pdf', format='pdf')
```

# Asymptotic null distribution of PET test

# Type I Error

```python
# Set seed
np.random.seed(seed=13)
```

```python
# Test parameters
N_rep = 500
level = 0.05
```

```python
# Initialize Type-I error dataframe
n_range = np.arange(100,500,100)
alpha_range = np.arange(0.1,0.5,0.1)
type_one_error = np.zeros((len(n_range),len(alpha_range)))
```

```python
# Run simulations
for k in tqdm(range(N_rep)):
    time.sleep(0.01)
    for i in range(len(n_range)):
        for j in range(len(alpha_range)):
            A = dg.sample_null_adj_mat(n_range[i], alpha_range[j])
            type_one_error[i,j] += st.PET_test(A, level)['reject']
type_one_error = type_one_error/N_rep
```

```python
# Construct Type I error dataframe
error_df = pd.DataFrame(type_one_error, n_range, np.round(alpha_range,2))
error_df
```
