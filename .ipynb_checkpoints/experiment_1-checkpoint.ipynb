# Case 1: Asymptotic null distribution of degree test

```python
# Data generation parameters
n = 200
alpha = 0.1
```

```python
# Test parameters
N_rep = 500
level = 0.05
```

```python
# Empirical null distribution
np.random.seed(seed=13)
test_stats = []
for i in tqdm(range(N_rep)):
    time.sleep(0.01)
    # Sample an adjacency matrix under the null
    A = dg.sample_null_adj_mat(n, alpha)
    # Compute degree test statistic
    t = st.degree_test(A, level)['test_stat']
    test_stats.append(t)
```

```python
# Plot histogram
degree_test = pd.DataFrame({'test_stat': test_stats})
ax = sns.histplot(data=degree_test, x='test_stat', stat='density')
xx = np.arange(-4,+4,0.001)
norm_pdf = ss.norm.pdf(xx)
ax.plot(xx,norm_pdf, 'r', lw=2)
ax.set_xlabel('degree test statistic')
ax.set_ylabel('density')
# Save figure to file
plt.savefig('exp1_case1_degree.pdf', format='pdf')
```

# Case 1: Asymptotic null distribution of PET test

```python
# Data generation parameters
n = 200
alpha = 0.1
```

```python
# Test parameters
N_rep = 500
level = 0.05
```

```python
# Empirical null distribution
np.random.seed(seed=13)
test_stats = []
for i in tqdm(range(N_rep)):
    time.sleep(0.01)
    # Sample an adjacency matrix under the null
    A = dg.sample_null_adj_mat(n, alpha)
    # Compute degree test statistic
    t = st.PET_test(A, level)['test_stat']
    test_stats.append(t)
```

```python
# Plot histogram
degree_test = pd.DataFrame({'test_stat': test_stats})
ax = sns.histplot(data=degree_test, x='test_stat', stat='density')
xx = np.arange(0,15,0.001)
norm_pdf = ss.chi2.pdf(xx, df=2)
ax.plot(xx,norm_pdf, 'r', lw=2)
ax.set_xlabel('PET test statistic')
ax.set_ylabel('density')
# Save figure to file
plt.savefig('exp1_case1_PET.pdf', format='pdf')
```
